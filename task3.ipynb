# 1) Importing Essential Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# 2) Read Original Image

```python
img=cv2.imread('dog.jpg')
```

```python
plt.imshow(img)
```

# 3) Read  Image in grayscale fromat

```python
img=cv2.imread('dog.jpg',0)
```

```python
plt.imshow(img, cmap='gray')
```

# 4) Invert image

```python
img.max()
```

```python
inv=255-img
```

```python
plt.imshow(inv,cmap='gray')
```

# 5) Apply Gaussian Blur

```python
blur=cv2.GaussianBlur(inv,ksize=(21,21),sigmaX=0,sigmaY=0)
```

```python
plt.imshow(blur,cmap='gray')
```

# 6)Then the final step is to invert the blurred image, then we can easily convert the image into a pencil sketch

```python
dodge=lambda image, mask: cv2.divide(image,255-mask,scale=256)
```

```python
blended=dodge(img,blur)
```

```python
plt.imshow(blended , cmap='gray')
```
